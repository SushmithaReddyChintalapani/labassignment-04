Data Preparation and Cleaning
Data Cleansing means the process of identifying incorrect, incomplete, inaccurate, irrelevant, or missing pieces of data and then modifying, replacing, or deleting them as needed. Data Cleansing is considered as the basic element of Data Science.

```
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set_style("darkgrid")
matplotlib.rcParams["font.size"] = 14
matplotlib.rcParams["figure.facecolor"] = '#00000000'
```

```
# represernting the data by a pie chart to understand the larger data easily
plt.figure(figsize=(12,6))
plt.title("Percentation of Netflix  movie Titles that are either Movies or TV Shows")
g = plt.pie(netflix_df.type.value_counts(),explode=(0.025,0.025), labels=netflix_df.type.value_counts().index, colors=['red','orange'],autopct='%1.1f%%', startangle=180)
plt.show()
```

 Quantity of new movies and TV shows uploaded to Netflix over the years. Given that we are curious as to when Netflix first made the title available to subscribers, we will create a new category labeled "year added" to display a year associated with the date values in the "date added" columns.

```
netflix_movies_df = netflix_df[netflix_df.type.str.contains("Movie")]
```

```
netflix_shows_df = netflix_df[netflix_df.type.str.contains("TV Show")]
```

```
netflix_year_df = netflix_df.year_added.value_counts().to_frame().reset_index().rename(columns={"index": "year", "year_added":"count"})
netflix_year_df = netflix_year_df[netflix_year_df.year != 2020]
```

```
# year and date where movies and tv shows were added
netflix_year_df
```

```
movies_year_df = netflix_movies_df.year_added.value_counts().to_frame().reset_index().rename(columns={"index": "year", "year_added":"count"})
movies_year_df = movies_year_df[movies_year_df != 2022]
```

```
movies_year_df
```

```
# top 40 shows mostly watched in the netflix
filtered_genres = netflix_df.set_index('title').listed_in.str.split(', ', expand=True).stack().reset_index(level=1, drop=True);

plt.figure(figsize=(8,11))
g = sns.countplot(y = filtered_genres, order=filtered_genres.value_counts().index[:20])
plt.title('Top 40 Genres on Netflix')
plt.xlabel(' Genres')
plt.ylabel('Titles')
plt.show()
```

```
# top rated movies and tv shows in the netflix during the years we have considered

order =  ['G', 'TV-Y', 'TV-G', 'PG', 'TV-Y7', 'TV-Y7-FV', 'TV-PG', 'PG-13', 'TV-14', 'R', 'NC-17', 'TV-MA']
plt.figure(figsize=(15,7))
g = sns.countplot(netflix_df.rating, hue=netflix_df.type, order=order, palette="pastel");
plt.title("Ratings for Movies & TV Shows")
plt.xlabel("Rating based on recent watch list ")
plt.ylabel("Total Count of the Movies and Tv Shows")
plt.show()
```

```
filtered_directors = netflix_df[netflix_df.director != 'No Director'].set_index('title').director.str.split(', ', expand=True).stack().reset_index(level=1, drop=True)
sns.countplot(y = filtered_directors, order=filtered_directors.value_counts().index[:10], palette='Greens')
plt.show()
```
